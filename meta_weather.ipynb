```python
import requests 
import csv 
```

```python
response = requests.get('https://www.metaweather.com/api/location/1521894/')
```

```python
weather = response.json()
weather["consolidated_weather"]
```

```python
with open('meta_weather.csv', 'w', newline='') as f :
    writer = csv.DictWriter(f, fieldnames=['the_temp', 'humidity', 'visibility', 'air_pressure', 'wind_speed', 'datetime'])
    writer.writeheader()

    for w in weather["consolidated_weather"] :
        tempreture = w ['the_temp']
        humidity =  w['humidity']
        visibilty = w['visibility']
        air_pressure = w['air_pressure']
        wind_speed = w['wind_speed']
        datetime =  w['created']

        writer.writerow( {
                'the_temp' :tempreture,
                'humidity' :humidity,
                 'visibility' :visibilty,
                  'wind_speed' :wind_speed,
                  'air_pressure' : air_pressure,
                   'datetime' :datetime
            } )
```
